```python
import requests
import datetime as dt
import pandas as pd
import pyarrow


load_date = datetime.date.today() 

response = requests.get(
    "https://api.open-meteo.com/v1/forecast",
    params = {
        "latitude": 53.9,
        "longitude": 27.5667,
        "hourly": ["temperature_2m", "snowfall", "snow_depth"],
        "start_date": "2025-12-20",
        "end_date": load_date
    },    
)

response.raise_for_status()
order_data_dict = response.json()

    
```

```python
filename = f"api_minsk_weather_{load_date}.csv"

df = pd.json_normalize(order_data_dict)

if len(df) > 0:
    df.to_csv(filename, sep=';')
    
```
